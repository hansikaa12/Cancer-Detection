```python
cat = []
num = []

for i in df.columns:
    if df[i].dtype == 'O':  # Correct way to check column dtype
        cat.append(i)
    else:
        num.append(i)
```

```python
cat
```

```python
for i in num:
    fig, ax = plt.subplots(1,2)
    ax[0].hist(df[i],color = 'c')
    ax[1].boxplot(df[i])
    plt.suptitle(i)
    plt.show()
```

```python
# iqr winsorization --
# replace lower outliers with lower fence
# replace upper outliers with upper fence


for i in num:
    q1 = df[i].quantile(0.25)
    q3 = df[i].quantile(0.75)
    iqr = q3 - q1
    lf = q1 - 1.5*iqr
    uf = q3 + 1.5*iqr
    df.loc[df[i]<lf,i] = lf
    df.loc[df[i]>uf,i] = uf
```

```python
import re

def convert_interval_to_numeric(value):
    match = re.findall(r"[-+]?\d*\.\d+|\d+", str(value))  # Extract numbers from the interval
    if len(match) == 2:
        return (float(match[0]) + float(match[1])) / 2  # Compute midpoint
    return np.nan  # If no valid numbers, return NaN

for col in df.columns:
    if df[col].dtype == 'O':  # Check object (categorical) columns
        df[col] = df[col].apply(convert_interval_to_numeric)  # Convert intervals to numeric

df.fillna(df.median(), inplace=True)  # Fill missing values after conversion
```
